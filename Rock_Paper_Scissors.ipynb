```python
import random

print("Welcome to 'Rock, Paper, Scissors' - my game!")

while True:
    steps = input("Please select the steps of the game. (Note: Please type an odd number): ")

    if steps.isdigit():
        int_steps = int(steps)
        if int_steps % 2 == 0:
            print("Please type an odd number.")
        else:
            print("The steps of the game:", int_steps)
            break
    else:
        print("Invalid input.")

com_score = 0
your_score = 0

while int_steps > 0:
    your_R_P_S = input("Rock, Paper, Scissors (r or p or s): ")
    words = ("r", "p", "s")
    com_R_P_S = random.choice(words)

    print(f"Your choice: {your_R_P_S}, Computer's choice: {com_R_P_S}")

    if your_R_P_S == com_R_P_S:
        print("Draw!")
    elif (your_R_P_S == "r" and com_R_P_S == "s") or (your_R_P_S == "p" and com_R_P_S == "r") or (your_R_P_S == "s" and com_R_P_S == "p"):
        print("You win!")
        your_score += 1
    else:
        print("Computer wins!")
        com_score += 1

    int_steps -= 1

print("Your score:", your_score)
print("Computer score:", com_score)

if your_score > com_score:
    print("You win! Good game.")
elif your_score == com_score:
    print("You draw. Good game.")
else:
    print("You lose. Good game.")
```

```python
import random

class Player:
    def __init__(self, name):
        self.name = name
        self.moves = ["rock", "paper", "scissors"]
        self.score = 0

    def choose_move(self):
        return random.choice(self.moves)

def determine_winner(player1, player2):
    if player1 == player2:
        return "It's a draw!"
    elif (player1 == "rock" and player2 == "scissors") or \
         (player1 == "paper" and player2 == "rock") or \
         (player1 == "scissors" and player2 == "paper"):
        return f"{player1.capitalize()} wins!"
    else:
        return f"{player2.capitalize()} wins!"

def play_game():
    player_name = input("Enter your name: ")
    human_player = Player(player_name)
    computer_player = Player("Computer")

    num_moves = int(input("Enter the number of moves you want to predict: "))

    for _ in range(num_moves):
        human_move = input("Enter your move (rock, paper, or scissors): ").lower()
        if human_move not in human_player.moves:
            print("Invalid move! Please choose rock, paper, or scissors.")
            continue

        computer_move = computer_player.choose_move()
        print(f"{computer_player.name}'s move: {computer_move}")

        result = determine_winner(human_move, computer_move)
        print(result)

        if "win" in result:
            human_player.score += 1
        elif "draw" not in result:
            computer_player.score += 1

    print(f"\nFinal Scores:\n{human_player.name}: {human_player.score}\n{computer_player.name}: {computer_player.score}")

if __name__ == "__main__":
    play_game()
scissors
```
